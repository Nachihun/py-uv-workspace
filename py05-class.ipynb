````python
class Person:
    def __init__(self, name, age, job):
      self.name = name   # 매개변수로 받어서 멤버변수에 값을 할당한다.
      self.age = age     # self.멤버변수 = 매개변수
      self.job = job


    def eat(self):
        print('eat')


    def walk(self):
        print('walk')
````

````python
hong = Person('홍길동', 30, '학생')
hong.eat()
hong.walk()
print(hong.name, hong.age, hong.job)
````

````python
lee = Person('이영희', 25, '개발자')
lee.eat( )
lee.walk()
print(lee.name, lee.age, lee.job)
````

````python
class FourCal:


  def __init__(self, a, b):
    print('init')
    self.a = a   # 멤버변수는 같은 클래스 내 다른 메소드에서 호출해서 참조가능하다.
    self.b = b


  def add(self):
    # 클래스의 멤버변수를 메소드를 호출해서 사용할 수 있다.
    return self.a + self.b


  def sub(self):
    return self.a - self.b


  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a // self.b
````

````python
fc = FourCal(4, 2)   # 생성자는 객체의 초기값을 할당한다.
````

````python
print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

### decorator

````python
def t(a):
    return a
````

````python
t(3)
````

````python
def x(fun):   # 매개변수로 함수를 받을 수 있고
    def y(x):
        return fun(x+1)
    return y   # 리턴값으로 함수를 넘길 수 있다., 외부에서 inner 함수 사용하기 위해 y함수 리턴
````

````python
m = x(t)
print(m.__name__)
print(type(m))

m(5)
````

````python
x(t)(5)    # m = x(t)
           # m(5)      =>  두 문장을 한 문장으로 표기
````
