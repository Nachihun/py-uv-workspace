```python
number = input("정수를 입력해 주세요. > ")
sum_numbers = 0

for i in number:
    sum_numbers += int(i)

if int(number)%2 == 0 and sum_numbers%3 == 0:
    print("입력한 숫자 {}은 짝수이고, 각 자리수 합은 3의 배수 입니다.".format(number))
elif int(number)%2 != 0 and sum_numbers%3 == 0:
    print("입력한 숫자 {}은 홀수이고, 각 자리수 합은 3의 배수 입니다.".format(number))
elif int(number)%2 == 0 and sum_numbers%3 != 0:
    print("입력한 숫자 {}은 짝수이고, 각 자리수 합은 3의 배수가 아닙니다.".format(number))
else:
    print("입력한 숫자 {}은 홀수이고, 각 자리수 합은 3의 배수가 아닙니다.".format(number))
```
