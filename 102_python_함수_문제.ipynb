[문제1]  
문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하세요.    
make_url("naver")

[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
def make_list(string):
    my_list = []

    for data in string:
        my_list.append(data)
    return my_list
```

```python
make_list("abcd")
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pick_even(items):
    result = []

    for item in items:
        if item % 2 == 0:
            result.append(item)
    return result

```

```python
pick_even([3, 4, 5, 6, 7, 8])
```

콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]

```python
def convert_int(str_int):
    s = str_int.replace(',', '')
    return int(s)


```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
max(10, 2, 5)

[출력결과]  
10

```python
def print_max(a, b, c):
    max_value = 0
    if a > max_value:
        max_value = a
    elif b > max_value:
        max_value = b
    else:
        max_value = c
```

```python
max(10, 2, 5)
```

2 와 5의 배수가 아닌 합계 :  16



```python
def resultr(*data):
    count = 0
    total = 0

    for i in data:
        if i % 2  != 0 and i % 5 != 0:
            count += 1
            total += i
    
    print("2와 5의 배수가 아닌 갯수 : ", count)
    print("2와 5의 배수가 아닌 합계 : ", total)
    # return count
    # return total
```
