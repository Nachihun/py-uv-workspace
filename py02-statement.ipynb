## 반복문 : while
반복문에서 else는 반복문이 완벽하게 끝났을 때 실행된다.  
대부분의 프로그래밍 언어는 while과 for을 지원한다.  
else
  - if ~ else    
  - while ~ else  
  - try ~ except ~ else ~ finally


while 조건식:

    수행할 문장

else:

    수행할 문장


```python
# while문이 완벽이  돌았으므로 end가 출력된다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
else:  # 옵션으로 생략 가능
  print('end')
```

```python
# while문을 강제적으로 빠져나왔기 때문에 else영역을 실행하지 않는다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
  if a==5:
    break # 반복문을 강제적으로 빠져나올때 사용한다. 즉 반복문을 중단시킨다
else:
  print('end')
```

```python
#  continue는 다음 루프(조건식)로 바로 넘어간다.
# 반복문을 중단하지는 않으므로 else문이 실행된다.
a = 10 #5
while a > 5:
  a -= 1
  if a % 2 == 0 :
    continue   # while 조건식으로 이동
  print(a) # if의 조건이 False일 때 실행
else:
  print('end')
```

## enumerate 함수
 - enumerate()함수는 인덱스와 객체를 동시에 저장할 수 있는 특징을 가진 for문을 좀더 효율적으로 사용하기 위한 함수이다.
 - enumerate()함수는 반복가능객체(list, tuple, str)을 입력으로 받아 인덱스를 포함하는 enumerate 객체를 리턴한다.
 - enumerate([순서자료형], [시작값=0])의 형식을 가지며 반복가능객체은 반복자가 가능한 객체가 입력되고 시작값은 입력시 해당수 부터 증가하며, 생략했을 경우 기본값은 0 이다.


```python
for k in enumerate(['apple', 'python', 'microsoft']):  # 데이터가 저장되어있는 인덱스 값도 같이 리턴해줌.
    print(k)
```

```python
for k, v in enumerate(['apple', 'python', 'microsoft']):  # 데이터가 저장되어있는 인덱스 값도 같이 리턴해줌.
    print(k, v)  # 인덱스와 같이  보여줄때는 튜플로 리턴해주나 따로 받으면 값으로 따로 리턴됨.
```

```python
# enumerate(iterable, start) =>  start은 생략이 가능하고 생략하면 defaul가 0이다.
for k, v in enumerate(['apple', 'python', 'microsoft'], 1):  # 시작하는 인덱스 값을 1로 지정해줌.
    print(k, v)
```

```python
a = {'one':10, 'two':20}
for k in enumerate(a):
  print(k)
```

```python
a = {'one':10, 'two':20}
for i, k in enumerate(a):
  print(i, a[k])
```

```python
a = {'one':10, 'two':20}
for _, k in enumerate(a):  # 인덱스 값을 안가져 올 때는 _(언더바) 입력해야 됨.
  print(a[k])
```

## List Comprehension
- List 정의시 값을 정하지 않고 호출시 리스트 내의 값들이 처리되도록 구성
    - A = [표현식 for 항목 in 반복가능객체(iterable)]
    - A = [표현식 for 항목 in 반복가능객체(iterable) if 논리식]
    - A = [표현식 for 항목1 in 반복가능객체(iterable)1 for 항목2 in 반복가능객체(iterable)2 if 논리식]


```python
s = []
for x in range(10):
    s.append(x**2)
print(s)
```

```python
s = [x**2 for x in range(10)]  # for x in range(10):
                               #    s.append(x**2)
print(s)                       # print(s)
                               # for문 먼저 실행하고 결과값이 x에 의해 리스트에 입력됨.
```

```python
s = []
for x in range(10):
    if x % 2 ==0:
        s.append(x**2)
print(s)
```

```python
s = [x**2 for x in range(10) if x % 2 == 0]  # 리스트에 저장될 값을 for문 왼쪽에 입력해준다. 딕셔너리 형태로 구할 때는 {}중괄호 사용
print(s)
```

```python
s = []
for x in range(5):
    if x % 2 ==0:
        s.append(x)
    else:
        s.append(-x)
print(s)
```

```python
s = [x if x%2 == 0 else -x for x in range(5)]
print(s)
```

```python
s = []
for i in range(1, 3):  # 1, 2
    for j in range(4, 8):  # 4, 5, 6, 7
        # print('({}, {})'.format(i,j))
        s.append((i,j))
print(s)
```

```python
s = [(i, j) for i in range(1, 3) for j in range(4, 8)]
print(s)
```

```python
s = []

cnt = [1, 1, 2, 1]    # 수량
prod = [500, 1000, 400, 600]  # 단가  

for i, j in zip(cnt, prod):
    s.append(i*j)
# print(s)

print(sum(s))
```

### Generator Expression
 - 파이썬에는 tuple(immutable)에는 comprehension이 존재하지 않고

 - 값을 하나씩 “생성”하는 반복자(iterator). 메모리를 덜 사용하며, **지연 평가(Lazy Evaluation)**가 특징이다.
 - generator를 호출하면 next 메소드를 통해 하나씩 호출하며 StopIteration 이 발생하면 종료된다.

반복문에서의 yield 사용
  - for문 내부에서는 자동으로 next()를 호출하고, StopIteration이 발생하면 반복 종료합니다.


```python
def count_up_t(n):
    i = 1
    while i <= n:
        yield i
        i += 1
        # print(i)

data = count_up_t(3)
print(data)
```

만약 1,000만 개의 데이터 중 연산 비용이 아주 큰 작업(예: time.sleep이 포함되거나 복잡한 수학 연산)을 수행하다가, 원하는 조건(예: 3의 배수)을 가진 첫 번째 결과만 찾고 싶을 때 두 방식의 차이

```python
import time


# 무겁고 오래 걸리는 연산을 시뮬레이션하는 함수
def heavy_processing(n):
    # 실제로는 아주 복잡한 연산이 일어난다고 가정합니다.
    return n * 2


# 1. 조기 평가 (Eager Evaluation) 방식
def eager_approach(data):
    # 조건과 상관없이 1,000만 개 데이터 전체에 대해 heavy_processing을 먼저 '전부' 실행함
    result = [heavy_processing(x) for x in data]  # 리스트로 표현식
   
    # 그 후, 필요한 조건을 찾음
    for val in result:
        if val > 0 and val % 3 == 0:
            return val


# 2. 지연 평가 (Lazy Evaluation) 방식
def lazy_approach(data):
    # 이 시점에는 아무런 연산도 하지 않음 (계산식만 준비)
    result_generator = (heavy_processing(x) for x in data)  # 제너레이터 표현식
   
    # 필요한 순간(next)에만 heavy_processing을 하나씩 실행함
    for val in result_generator:
        if val > 0 and val % 3 == 0:
            return val


# 테스트 데이터 (0부터 9,999,999까지)
data_list = range(10000000)


# [비교 시작]
# 1) 조기 평가는 1,000만 번의 heavy_processing을 모두 수행하느라 엄청난 시간이 걸림
# eager_approach(data_list)
start_time = time.time()
print(f"즉시 평가 결과: {eager_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")


# 2) 지연 평가는 단 4번(x=0,1,2,3)만 연산하고 x=3일 때 6(3의 배수)을 반환하며 즉시 종료!
start_time2 = time.time()
print(f"지연 평가 결과: {lazy_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time2:.6f}초")
```

예제 1: 뉴스 기사 요약용 키워드 필터링기 (Comprehension 활용)


- 목표: 주어진 뉴스 본문 리스트에서 특정 키워드를 포함한 문장만 필터링


```python
news_articles = [
    "ChatGPT is transforming the AI world.",
    "Bitcoin price hits a new high.",
    "Python dominates machine learning.",
    "Weather forecast shows heavy rain.",
]


keywords = ['AI', 'Python', 'machine learning']

filtered = [ article for article in news_articles if any(k in article for k in keywords) ]  # any(True, False, False) : 하나만 True여도 True값 리턴 

print(filtered)
```
